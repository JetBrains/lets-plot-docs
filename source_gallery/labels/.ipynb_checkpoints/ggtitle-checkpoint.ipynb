# ggtitle()

Build a plot with `ggtitle()`.

```python
import pandas as pd

from lets_plot import *
LetsPlot.setup_html()
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/HIL-HK/lets-plot-examples/master/data/mpg.csv')
```

```python
ggplot(df, aes(x='cty', y='hwy')) + \
    geom_point() + \
    ggtitle('New Plot Title')
```
